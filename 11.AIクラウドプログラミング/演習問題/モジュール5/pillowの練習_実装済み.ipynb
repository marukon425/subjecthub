## 新規画像の作成

Image.new('RGB', サイズ, 色)

色:(赤, 緑, 青)

問 幅400px、高さ200px、色がR:230、G:230、B:230の画像を作成してください

## 部分切り抜き  
https://note.nkmk.me/python-pillow-image-crop-trimming/

画像"target.jpg"から次の範囲を切り抜いて表示してください  
左:100、上:200、右:400、下:700

```python
# 画像ファイルのオープン ... 画像オブジェクトの生成
image_in = Image.open('target.jpg')

# 切り抜き
image_crop = image_in.crop((100, 200, 400, 700))

# 画像を表示
image_crop.show()
```

## リサイズ  
https://note.nkmk.me/python-pillow-image-resize/

問 画像"target.jpg"を幅256px、高さ256pxにリサイズし、表示してください

```python
# 画像ファイルのオープン ... 画像オブジェクトの生成
image_in = Image.open('target.jpg')

# リサイズ
image_resize = image_in.resize((256, 256))

# 画像を表示
image_resize.show()
```

問 画像"target"の幅と高さの縦横比を保ったまま0.3倍にリサイズし、表示してください

```python
# 画像ファイルのオープン ... 画像オブジェクトの生成
image_in = Image.open('target.jpg')

# リサイズ
rate = 0.3
image_resize = image_in.resize( (int(image_in.width * rate), int(image_in.height * rate)) )

# 画像を表示
image_resize.show()
```

## 線の描画  
https://note.nkmk.me/python-pillow-imagedraw/

問 新規画像を作成し、画像の範囲内に線を描画してください  
線について  
・開始x:100、開始y:50、終了x:200、終了y:100  
・塗りつぶしの色R:255、G:0、B:0  
・線の太さ:10

```python
from PIL import Image
from PIL import ImageDraw

# 画像の新規作成
image_new = Image.new('RGB', (400, 200), (230, 230, 230))

# 画像オブジェクトに対する描画オブジェクトの生成
draw = ImageDraw.Draw(image_new)

# 線の描画
draw.line((100, 50, 200, 100), fill=(255, 0, 0), width=10)

# 画像を表示(描画された状態になっている)
image_new.show()
```

問 画像"target.jpg"を開き、画像の範囲内に線を描画してください  
線について  
・開始x:100、開始y:50、終了x:200、終了y:100  
・塗りつぶしの色R:255、G:0、B:0  
・線の太さ:10

```python
# すでにある画像に対しても描画できる
from PIL import Image
from PIL import ImageDraw

# 画像ファイルのオープン ... 画像オブジェクトの生成
image_in = Image.open('target.jpg')

# 画像オブジェクトに対する描画オブジェクトの生成
draw = ImageDraw.Draw(image_in)

# 線の描画
draw.line((100, 50, 200, 100), fill=(255, 0, 0), width=10)

# 画像を表示(描画された状態になっている)
image_in.show()
```
